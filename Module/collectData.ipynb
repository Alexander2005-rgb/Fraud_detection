```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Boxplot to visualize outliers
plt.figure(figsize=(6,4))
sns.boxplot(x=df['policy_annual_premium'])
plt.title("Boxplot of Policy Annual Premium")
plt.show()

# Calculate IQR
Q1 = df['policy_annual_premium'].quantile(0.25)
Q3 = df['policy_annual_premium'].quantile(0.75)

IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Function to visualize distribution and probability plot
def plot_data(feature):
    
    plt.figure(figsize=(12,5))
    
    plt.subplot(1,2,1)
    sns.histplot(feature, kde=True)
    plt.title("Distribution Plot")
    
    plt.subplot(1,2,2)
    stats.probplot(feature, dist="norm", plot=plt)
    plt.title("Probability Plot")
    
    plt.show()

# Visualize before transformation
print("Before Log Transformation")
plot_data(df['policy_annual_premium'])

# Apply log transformation to handle outliers
df['policy_annual_premium_log'] = np.log(df['policy_annual_premium'])

# Visualize after transformation
print("After Log Transformation")
plot_data(df['policy_annual_premium_log'])
```
